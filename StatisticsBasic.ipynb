Q1. Explain the different types of data (qualitative and quantitative) and provide examples of each. Discuss nominal, ordinal, interval, and ratio scales

 ->Data is broadly classified into two types:

Qualitative Data (Categorical): Descriptive data that cannot be measured numerically.

Nominal: Categories without a specific order (e.g., gender, color).

Ordinal: Categories with a specific order but no consistent difference (e.g., rankings, satisfaction level).

Quantitative Data (Numerical): Data that can be measured and expressed numerically.

Interval: Ordered with meaningful differences but no true zero (e.g., temperature in Celsius).

Ratio: Like interval, but has a true zero (e.g., height, weight, age).

Q2. What are the measures of central tendency, and when should you use each? Discuss the mean, median, and mode with examples and situations where each is appropriate.

Answer 2:

Mean: The average of the dataset. Use when data is symmetric without outliers.Example: Mean income in a city.

Median: Middle value when data is sorted. Best for skewed distributions or outliers.Example: Median house price.

Mode: Most frequently occurring value. Used for categorical data.Example: Most common shoe size sold.

Q3. Explain the concept of dispersion. How do variance and standard deviation measure the spread of data?

Answer 3:Dispersion refers to how spread out the data is.

Variance: Average of squared differences from the mean.

Standard Deviation (SD): Square root of the variance.
A higher SD means more spread in the data.

Q4. What is a box plot, and what can it tell you about the distribution of data?

Answer 4:A box plot is a graphical summary showing minimum, Q1, median, Q3, and maximum. It identifies skewness and outliers easily.

Q5. Discuss the role of random sampling in making inferences about populations.

Answer 5:Random sampling ensures each member of the population has an equal chance of selection, reducing bias and enabling generalization from sample to population.

Q6. Explain the concept of skewness and its types. How does skewness affect the interpretation of data?

Answer 6:Skewness measures the asymmetry of the data distribution.

Positive Skew: Tail on the right; mean > median.

Negative Skew: Tail on the left; mean < median.
Skewness affects how we interpret the mean and variance of the dataset.

Q7. What is the interquartile range (IQR), and how is it used to detect outliers?

Answer 7:IQR = Q3 - Q1. It measures the middle 50% spread. Outliers are typically data points 1.5 * IQR above Q3 or below Q1.

Q8. Discuss the conditions under which the binomial distribution is used.

Answer 8:Used when:

Fixed number of trials

Each trial has two outcomes (success/failure)

Trials are independent

Probability of success is constant
Example: Flipping a coin 10 times.

Q9. Explain the properties of the normal distribution and the empirical rule (68-95-99.7 rule).

Answer 9:A symmetric, bell-shaped distribution.

Mean = Median = Mode

Empirical Rule:

68% within 1 SD

95% within 2 SD

99.7% within 3 SD

Q11. Explain what a random variable is and differentiate between discrete and continuous random variables.

Answer 11:

Random Variable: Assigns numerical values to outcomes.

Discrete: Countable (e.g., number of students).

Continuous: Infinite values within range (e.g., height).

Q12. Provide an example dataset, calculate both covariance and correlation, and interpret the results.

Answer 12:Dataset: X = [2, 4, 6], Y = [1, 3, 5]

Covariance:

Mean X = 4, Mean Y = 3

Cov(X,Y) = [(2-4)(1-3)+(4-4)(3-3)+(6-4)(5-3)] / 2 = (4+0+4)/2 = 4

Correlation:

SD(X)=2, SD(Y)=2

Corr = Cov/(SDxSDy) = 4/(22) = 1
Interpretation: Strong positive linear relationship.
